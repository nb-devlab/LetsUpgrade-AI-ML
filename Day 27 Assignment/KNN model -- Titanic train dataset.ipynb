# Using the Titanic Dataset to implement KNN classification technique

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('./Day 24 Assignment/train.csv')

df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis=1, inplace=True)

df.isna().sum()
```

```python
# Using LabelEncoder on 'Sex' & 'Embarked' variable

le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df['Sex'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
df.columns
```

```python
# Split data into X and y for IDV and DV variables.

X = df.drop(['Pclass'], axis=1)
y = df['Pclass']
```

```python
# Using train_test_split to split data into train and test dataset

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
# Import KNN
from sklearn import neighbors
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=3)     # n_neighbors --> K value
```

```python
model = knn.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

Based on the train dataset, the output for test dataset model is classified accurately at 86.51%

```python
y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

A total of 267 observations are recorded in the matrix of which 231 are correct and the rest incorrect.
A k-value of 1 to 267 is taken to plot a graph of the same & its highest model accuracy to be recorded.

```python
def knn_model():
    l1 = []    # k value
    l2 = []    # model accuracy
    for i in range(1,268):
        knn = neighbors.KNeighborsClassifier(n_neighbors=i)
        model_acc = knn.fit(X_train,y_train).score(X_test, y_test)
        y_pred = knn.predict(X_test)
        l2.append(model_acc)
        l1.append(i)
        print('When the k value given is :', i)
        print('The model accuracy is :', model_acc)
        print('\n')
    
    plt.plot(l1, l2)
    plt.title('K-value vs Model Accuracy Score')
    plt.xlabel('K-value')
    plt.ylabel('Model Accuracy')
    plt.figure(figsize=(20,10))
    plt.show()
        
knn_model()
```

The model accuracy of 89.13% is highest at k-vakue of 1.
A general decline in model accuracy is observed therafter except for a few spikes in between.
