Chi Square Test

```python
from scipy.stats import chi2_contingency
```

```python
chitable = pd.crosstab(dataset1['Attrition'], dataset1['Education'])

chitable
```

```python
stats,p, dof, expected=chi2_contingency(chitable)
```

```python
print(stats,p)
```

p value > 0.05. Hence the Null Hypothesis is accepted.
There is no significant impact of Education on Attrition.

```python
chitable = pd.crosstab(dataset1['Attrition'], dataset1['JobLevel'])

chitable
```

```python
stats, p, dof, expected=chi2_contingency(chitable)
```

```python
print(stats,p)
```

p value > 0.05. Hence the Null Hypothesis is accepted. There is no significant impact of JobLevel on Attrition.

```python
from scipy.stats import mannwhitneyu
```

```python
stats, p = mannwhitneyu(dataset_AttritionYes['DistanceFromHome'], dataset_AttritionNo['DistanceFromHome'])
```

```python
print(stats,p)
```

p value > 0.05. Hence the H0 is accepted. There is no significant impact of DistanceFromHome on Attrition Yes and No.

```python
stats, p = mannwhitneyu(dataset_AttritionYes['Education'], dataset_AttritionNo['Education'])
```

```python
print(stats,p)
```

```python
print(stats,p)
```

p value < 0.05. Hence H0 is rejected. Therefore we can conclude that there is a difference of Age between Attrition Yes and Attrition No.
